```python
location_dic={}

contries = []
cities = []
#platforms = []


location_dic = {'country' : contries ,
               'city' : cities} 
               #'platform' : platforms

for i in location_list:
    if i[0] not in cities:
        cities.append(i[0])
        contries.append(i[1])
        #platforms.append(train.loc[i, 'platform'])
```
