## 호텔 종류 분류 
#### Full-service Hotel :  
- 흔히 아는 5, 6, 7성급 호텔,  
- 객실, 레스토랑, 회의/연회시설 등 다양한 부대시설  
- 비즈니스 고객 낯선 도시에 방문시 호텔 내부의 시설에 의존  
- feels like home 모토  
- 고객 편의와 동선을 고려한 시설과 서비스 제공, 높은 서비스 품질  
- 실내/외 수영장, 스파시설  
  
#### Selected-service Hotel :  
- 4성급 호텔, Full-service 호텔과 Buget hotel 사이에 위치한 카테고리  
- full-service 와 차이  
- 부대시설의 차이. 비즈니스 고객을 위한 회의나 연회시설 보유.  
- but 수백명이 참석하는 대형 컨퍼런스나 행사 유치에는 적합하지 않음  
- 간단한 헬스장, 투숙객을 위함 피트니스 센터, 수영장 ㄴㄴ  
- buget hotel 과 차이  
- 서비스와 부대시설 차이. 객실의 크기나 인테리어는 full-service 지향.  
- 부대시설을 이용할 의사는 없지만 고급스럽고 편리한 객실, 서비스  
  
#### Budget Hotel:  
- 중저가 호텔  
- 주로 단기체류 목적, 제한된 서비스와 합리적인 가격  
- B&B Bed and breakfast  
- 작은 객실크기, 제한된 부대시설, 사이즈 간소화, 합리적 소비  
- 기본적인 레스토랑, 간단한 비즈니스 업무 할 수 있는 데스크탑  
- 고객안전 보장, 기본적인 서비스 제공  
  
#### Longstay, Apartment, Extend stay  
- 30일 이상 장기체류를 목적으로 하는 고객 대상  
- 객실내부 아파트와 같이 간단한 가전제품과 주방시설,  
- 호텔에서 제공하는 기본적인 서비스  
- (콜센터,룸서비스,레스토랑,헬스/수영장)
