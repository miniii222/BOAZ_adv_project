### 1. 원본 데이터 row만 90만개. 3만개로 잘라서 분석해보자. 

```python
data = item.iloc[:30000,:]
```

```python
data.shape
```

```python
data.properties[:5]
```

```python
vectorizer = CountVectorizer(min_df = 0, ngram_range = (1,3))
# min_df 단어장에 추가되기 위한 최소 빈도 (토큰이 나타날 최소 문서 개수)
# ngram_range: n 범위 (min_n, max_n)
print(vectorizer)
```

```python
pipeline = Pipeline([('vect',vectorizer)])
```

```python
hotel_features = pipeline.fit_transform(data.properties) # learn vocab dictionary(token) from `corpus`
```

```python
hotel_features # 파씽 기준이 잘못 설정됐다! 파씽 기준 다시 설정하기 
```
