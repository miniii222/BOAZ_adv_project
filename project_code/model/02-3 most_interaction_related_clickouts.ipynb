## 마지막 clickout item에 해당하는 row들의 reference만 고려해 counts 순위로 ranking정하기

## CASE 1 
clickout 개수가 1개인 session 
* df_mono: 한 session내에서 clickout 횟수가 1번인 session들의 df
* df_target: df_mono의 target row

```
def mono(df):
  s = (df.loc[df.action_type == 'clickout item'].groupby('session_id').action_type.count() == 1)
  mono_sess_index = s.index[s].to_list() # return session_id index
  df_mono = df.loc[df.session_id.isin(mono_sess_index)]
  
  return df_mono
```

```
df_mono = mono(df_test)
```

```
df_expl = f.explode(df_target, "impressions")
```

```
df_expl
```

```
# reference에서 letter, nan 제거
def only_number(string):
  try:
    int(string)
    return int(string)
  except ValueError:
    return np.nan
```

```
df_mono['reference_int'] = df_mono.reference.apply(only_number)
```

```
df_mono
```

```
def reference_count(df):
  multi = df.groupby(df.session_id).reference_int.value_counts() # na값 drop by default 
  df_out = pd.DataFrame(multi).rename(columns = {'reference_int':'ref_count'}).reset_index()
  return df_out
```

```
df_ref_count = reference_count(df_mono)
```

```
df_ref_count
```

```
df_expl
```

```
def ref_recommendation(df_expl, df_ref_count):
    """Calculate recommendations based on the number of being referred.
    The final data frame will have an impression list sorted according to the number of being referred item.
    :param df_expl: Data frame with exploded impression list
    :param df_pop: Data frame with session, items and number of being referred
    :return: Data frame with sorted impression list according to the number of being referred in df_ref_count.
    """
    GR_COLS =  ["user_id", "session_id", "timestamp", "step"]
    
    df_expl_clicks = (
        df_expl[GR_COLS + ["impressions"]]
        .merge(df_ref_count,
               left_on=["session_id", "impressions"],
               right_on=["session_id", "reference_int"],
               how="left")
    )

    df_out = (
        df_expl_clicks
        .assign(impressions=lambda x: x["impressions"].apply(str))
        .sort_values(GR_COLS + ["ref_count"],
                     ascending=[True, True, True, True, False])
    )

    df_out = f.group_concat(df_out, GR_COLS, "impressions")
    df_out.rename(columns={'impressions': 'item_recommendations'}, inplace=True)

    return df_out
```

```
df_out = ref_recommendation(df_expl, df_ref_count)
```

```
df_out
```

```
df_out.to_csv(path + 'mono.csv')
```

## CASE 2
clickout 개수가 2개 이상인 session 

*  df_many : 한 session내에서 clickout 횟수가 1번 이상인 session들의 df
* *주의* : clickout끼리 붙어있는 경우와 clickout끼리 떨어진 경우 구분해야한다.  [접근방법]  pure (session별 모든 reference) - previous (clickout한 reference) 

```
def many(df):
  s = (df.loc[df.action_type == 'clickout item'].groupby('session_id').action_type.count() != 1)
  many_sess_index = s.index[s].to_list() # return session_id index
  df_many = df.loc[df.session_id.isin(many_sess_index)]
  
  return df_many
```

```
df_many = many(df_test)
```

```
df_many_with_clickout = df_many.loc[(df_many.action_type == "clickout item") & (df_many.reference.notnull())] # clickout row 
```

```
df_many_no_clickout = df_test.iloc[df_many.index.difference(df_many_with_clickout.index)] # not clickout row
```

```
df_many_with_clickout['reference_int'] = df_many_with_clickout.reference.apply(only_number)
```

```
df_ref_count_no_clickout = reference_count(df_many_no_clickout)
```

```
df_ref_count_no_clickout
```

```
df_ref_count_with_clickout = reference_count(df_many_with_clickout)
```

```
# mapping two dataframes 
df_new = df_ref_count_no_clickout.merge(df_ref_count_with_clickout.drop('ref_count', axis = 1).rename(columns = {"reference_int" : "clickout"}), left_on = ['session_id', 'reference_int'], right_on = ['session_id', 'clickout'], how = 'left')
```

```
df_ref_count =  df_new.loc[df_new.reference_int != df_new.clickout]
```
